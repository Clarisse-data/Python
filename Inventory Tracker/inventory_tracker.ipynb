```python
# Craft Store Inventory Tracker Activity
inventory = []

# Add items to inventory
def add_items():
  name = input('Enter name of item: ')
  quantity = int(input('Enter quantity of item: '))
  inventory.append((name,quantity))

# Show items from inventory
def show_items():
  print('\nInventory List:')
  for item in inventory:
    print(f'Item: {item[0]}, Quantity: {item[1]}')
  print('\n')

# Remove items from inventory
def remove_items():
  remove_item_name = input('Enter item name to remove: ')
  for item in inventory:
    if item[0] == remove_item_name:
      inventory.remove(item)

# Update quantity of item in inventory
def update_items():
  update_item_name = input('Enter item name to update: ')
  update_item_quantity = input('Enter item quantity to update: ')
  for item in inventory:
    if item[0] == update_item_name:
      inventory.remove(item)
      inventory.append((update_item_name,update_item_quantity))
      break

# Requests input to add or finish with the inventory tracker, and display
print('Welcome to the Craft Store Inventory Tracker.')
while True:
  menu_choice = input('Do you want to add/remove/see/update/exit items: ').lower()
  if menu_choice == 'add':
    add_items()
  elif menu_choice == 'remove':
    remove_items()
  elif menu_choice == 'see':
    show_items()
  elif menu_choice == 'update':
    update_items()
  elif menu_choice == 'exit':
    break
  else:
    print('Invalid response, please try again.')
```
